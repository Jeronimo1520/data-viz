```python
import pandas as pd
```

```python
df = pd.read_csv('../tiendas_procesado.csv')
df = df.set_index('mes')
df.to_csv('tiendas_procesado_index_mes.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (15,3) #Cambiar globalmente
```

El siguiente grafico de lineas tiene su grid con lineas tipo dashed(-----)

```python
plt.style.use(
    #'ggplot'
    'bmh'
    #'grayscale'
)

ax = df.plot.line(
    figsize = (10, 5) #Ancho y alto de la figura en pulgadas, sobreescribe el global
    
)

#Personalizar cada eje del grid

ax.xaxis.grid(
    color='blue',
    alpha = 0.3 , #opacidad o transparencia
    linewidth = 0.5,
    #visible = False
)

ax.yaxis.grid(
    color = '#34eb98',
    linestyle = '-.'
)

ax
```

```python
import seaborn as sns
```

```python
plt.figure(
    figsize = (10,5)
) #metodo global

sns.set_style('ticks') #Estilos
sns.set_theme('poster') #Temas para un fin objetivo, diapositiva, paper, etc.

ax = sns.lineplot(
    data = df
)

plt.grid( #Metodo global
    visible = False,
    axis = 'x'
)

plt.grid( #Metodo global
    axis = 'y',
    color = 'red',
    alpha = 0.2
)

#Se puede aplicar lo mismo de matplotlib a seaborn para modificar los ejes
# ax.xaxis.grid(
#     color='blue',
#     alpha = 0.3 , #opacidad o transparencia
#     linewidth = 0.5,
#     #visible = False
# )
```

```python
import plotly.express as px
```

Nota: Recordar que plotly no es basado en matplotlib

```python
fig = px.line(
    df,
    width = 800, #Medidas en pixeles, no en pulgadas 
    height = 300,
    template = 'plotly_dark',
    #template = 'ggplot2'
)

fig.update_xaxes(showgrid = False)

fig.update_yaxes(
    gridwidth = 1,
    gridcolor = 'white',
    griddash = 'dot',
    minor_griddash = 'dot' #Cuadriculas auxiliares
)
```
