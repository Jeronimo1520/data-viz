```python
import pandas as pd
df = pd.read_csv('../tiendas_procesado.csv')
df = df.set_index('mes')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use(
    'ggplot'
    #'bmh'
    #'grayscale'
)

ax = df.plot.line(
    figsize = (15, 7.5), #Ancho y alto de la figura en pulgadas, sobreescribe el global
)

ax.set_title(
    'Ventas de 3 tipos de productos en el año 2015',
    fontdict = {
        'size' : 30,
        'weight' : 'bold', #bold, light, etc
    }
)

posiciones_ticks = np.arange(0,len(df))

ax.set_xticks(
    posiciones_ticks
)

ax.set_xticklabels(
    df.index.values,
    fontdict={
        'color':'blue',
        'name':'Times New Roman',
        'weight': 15,
        'weight' : 'bold'
        
    },
    rotation = 45
)

# ax.set_xticks(
#     posiciones_ticks[::3]
# )

# ax.set_xticklabels(
#     df.index.values[::3],
#     rotation = 45
# )

# ax.set_yticks(
#     [500000,1000000]
# )

# ax.set_yticklabels(
#     ['500000', '1000000'],
#     fontdict = {
#         'size':25
#     }
# )

# ax.set_xticks(
#     [0,1] #Posiciones
# )

# ax.set_xticklabels(
#     ['ENERO', 'FEBRERO'] #Etiquetas,
#     rotation = 45
# )

ax.set_xlabel(
    'Periodo',
    loc = 'center'
)

ax.set_ylabel(
    'Ventas en millones de dólares',
    loc = 'center',
    fontdict={
        'name':'Comic Sans MS',
        'color': 'red',
        'size' : 30
    }
)

plt.yticks( #Metodo global, en caso de no redefinir las etiq del eje Y 
    color = 'blue',
    name = 'Times New Roman'
)

plt.xticks( #Metodo global, en caso de no redefinir las etiq del eje Y 
    color = 'green',
    name = 'Times New Roman'
)


ax.legend(loc='upper left',prop={'size': 13})
```

```python
import seaborn as sns

plt.figure(
    figsize = (10,5)
) #metodo global


#Forma 1 TITULOS
plt.title('Ventas de 3 productos en el año 2015')
plt.xlabel('Periodo')
plt.ylabel('Ventas en millones de dólares')

sns.set_style('ticks') #Estilos
sns.set_theme('poster') #Temas para un fin objetivo, diapositiva, paper, etc.

ax = sns.lineplot(
    data = df
)

ax.set_xticks(
    [0,1]
)

ax.set_xticklabels(
    ['ENERO', 'FEBRERO'],
    rotation = 45
)

#Forma 2 TITULOS
ax.set_xlabel(
    'Periodo'
)

ax.set_ylabel(
    'Mi eje y',
     fontdict={ #Personalizacion igual a matplotlib
        'name':'Comic Sans MS',
        'color': 'red',
        'size' : 30
    }
)


#Forma 3 TITULOS
ax.set(
    xlabel = "Periodo",
    ylabel = 'Ventas en millones de dólares',
    title = 'Ventas de 3 productos en el año 2015'
)


plt.xticks(rotation=45) 
```

```python
fig = px.line(
    df,
    width = 800, #Medidas en pixeles, no en pulgadas 
    height = 500,
    #template = 'plotly_dark',
    template = 'ggplot2',
    title= 'Ventas de 3 tipos de producto por año',
#     labels = { 
#         'mes' : 'periodo',
#         'value': 'Ventas en millones de dolares',
#         'variable' : 'Tipo de producto'
#     }
#Primera forma
)

fig.update_layout(
    xaxis_title = 'PERIODO',
    yaxis_title = 'VENTAS EN MILLONES DE USD',
    font_color = 'blue ', #Cambios globales
    font_family = 'Courier New',
    legend_font_color = 'black',
    legend_title_font_color = 'green',
    title_font_color = 'red',
    title_font_family = 'Times New Roman'
    
)

fig.update_xaxes(
    tickangle = 45,
    tickvals = df.index.values,
    tickmode = 'array', #Arreglo de etiquetas
    title_font_family = 'Arial', #Cambio solo al eje 
    title_font_color = 'black',
    title_font_size = 26,
    tickfont = {
        'family':'Rockwell',
        'color':'brown',
        'size' : 10
    }
)

fig.update_yaxes(
    #nticks = 8, #Numero de etiquetas APROXIMADO en el eje Y 
    #tickvals = [500000,1000000]
    dtick = 250000 #Distancia
)

fig.show()
```
