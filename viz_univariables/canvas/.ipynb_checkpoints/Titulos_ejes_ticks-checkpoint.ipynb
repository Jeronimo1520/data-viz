```python
import pandas as pd
df = pd.read_csv('../tiendas_procesado.csv')
df = df.set_index('mes')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use(
    'ggplot'
    #'bmh'
    #'grayscale'
)

ax = df.plot.line(
    figsize = (10, 5), #Ancho y alto de la figura en pulgadas, sobreescribe el global
    title = 'Ventas de 3 tipos de productos en el año 2015'
)

ax.set_xlabel(
    'Periodo',
    loc = 'center'
)

ax.set_ylabel(
    'Ventas en millones de dólares',
    loc = 'center'
)
```

```python
import seaborn as sns

plt.figure(
    figsize = (10,5)
) #metodo global


#Forma 1
plt.title('Ventas de 3 productos en el año 2015')
plt.xlabel('Periodo')
plt.ylabel('Ventas en millones de dólares')

sns.set_style('ticks') #Estilos
sns.set_theme('poster') #Temas para un fin objetivo, diapositiva, paper, etc.

ax = sns.lineplot(
    data = df
)

#Forma 2
ax.set_xlabel(
    'Periodo'
)

ax.set_ylabel(
    'Mi eje y'
)


#Forma 3
ax.set(
    xlabel = "Periodo",
    ylabel = 'Ventas en millones de dólares',
    title = 'Ventas de 3 productos en el año 2015'
)


plt.xticks(rotation=45) 
```
