```python
import pandas as pd
df = pd.read_csv('../tiendas_procesado.csv')
df = df.set_index('mes')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use(
    'ggplot'
    #'bmh'
    #'grayscale'
)

ax = df.plot.line(
    figsize = (15, 7.5), #Ancho y alto de la figura en pulgadas, sobreescribe el global
)

ax.set_title(
    'Ventas de 3 tipos de productos en el año 2015',
    fontdict = {
        'size' : 30,
        'weight' : 'bold', #bold, light, etc
    }
)

posiciones_ticks = np.arange(0,len(df))

ax.set_xticks(
    posiciones_ticks
)

ax.set_xticklabels(
    df.index.values,
    fontdict={
        'color':'black',
        'weight': 10,
        
    },
    rotation = 45
)

ax.set_xlabel(
    'Periodo',
    size = 20,
    fontdict={
        'color':'black',
          'weight': 12,
    },
    loc = 'center'
)

ax.set_ylabel(
    'Ventas en millones de dólares',
    size = 18,
    loc = 'center',
    fontdict={
        'color': 'black',
        'size' : 12
    }
)

ax.annotate( #Permite hacer anotaciones y destacar algo en el grafico
    'Noviembre OJO!!',
    size = 20,
    xy = (10, 1250000), #Donde apunta la flecha
    xytext = (10, 1000000), #Donde apunta el cuerpo de la flecha
    color = 'grey',
    horizontalalignment = 'center',
    arrowprops={
        'width' : 1,
        'color' : 'black',
        'headwidth' : 5,
        'headlength' : 5
    }
)

ax.legend(
    title='Tipo de Producto', # El titulo,
    title_fontsize = 20,
    labels=['Lacteos', 'Carnes', 'Comida de Mar'], # Puedo redefinir el texto de mi leyenda
    loc='upper left', # Punto de referencia para posicionar mi leyenda
    bbox_to_anchor=(1, 1), # Porcentaje del eje X y Y donde quiero colocar mi leyenda tomando como referencia "loc"
#     title_fontproperties={
#         'size': 14
#     },
    fontsize=15, # Para los labels de la leyenda
    labelcolor=['red', 'blue', 'purple'], # Puedo configurar colores por cada uno de las etiquetas
#     alignment='left', # 'left', 'center', 'right': Alineación de todos los textos
    #ncol=3
)

```

```python
import plotly.express as px
fig = px.line(
    df,
    width = 800, #Medidas en pixeles, no en pulgadas 
    height = 500,
    #template = 'plotly_dark',
    template = 'ggplot2',
    title= 'Ventas de 3 tipos de producto por año',
#     labels = { 
#         'mes' : 'periodo',
#         'value': 'Ventas en millones de dolares',
#         'variable' : 'Tipo de producto'
#     }
#Primera forma
)

fig.update_layout(
    xaxis_title = 'PERIODO',
    yaxis_title = 'VENTAS EN MILLONES DE USD',
    font_color = 'blue ', #Cambios globales
    font_family = 'Courier New',
    legend_font_color = 'black',
    legend_title_font_color = 'green',
    title_font_color = 'red',
    title_font_family = 'Times New Roman'
    
)

fig.update_xaxes(
    tickangle = 45,
    tickvals = df.index.values,
    tickmode = 'array', #Arreglo de etiquetas
    title_font_family = 'Arial', #Cambio solo al eje 
    title_font_color = 'black',
    title_font_size = 26,
    tickfont = {
        'family':'Rockwell',
        'color':'brown',
        'size' : 10
    }
)

fig.update_yaxes(
    #nticks = 8, #Numero de etiquetas APROXIMADO en el eje Y 
    #tickvals = [500000,1000000]
    dtick = 250000 #Distancia
)

fig.add_annotation(
    x = '2015-11',
    y = 1300000,
    text = 'Noviembre!!!!',
    opacity = 0.5,
    font = {
        'size': 10,
        'color' : 'black'
    },
    showarrow = True,
    arrowsize = 1,
    arrowwidth = 2,
    arrowhead = 2 #es un ID del 0 al 8, que indica el tipo de flecha
)

fig.add_hline(
    y = 1300000,
    annotation_text = 'Umbral',
    annotation_position = 'bottom right',
    annotation_font_color = 'red',
    annotation_font_size = 10,
    line_dash = 'dot',
    opacity = 0.8,
    line_width = 1,
    line_color = 'red'
)

fig.update_layout(
    legend_title_text = 'Leyenda',
    legend = {
        'y' : 0.5,
        'x' : 1.05,
        'font': {
            'family' : 'Courier',
            'size' : 12,
            'color' : 'green'
        },
        'bgcolor' : 'LightGrey' ,
        'bordercolor' : 'Black',
        'borderwidth' : 2
    },
    
    
)

fig.show()
```
