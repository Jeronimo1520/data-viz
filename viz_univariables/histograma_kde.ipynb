# Histograma

Diagrama de barras que representa la distribución de una variable numérica.

Los datos se agrupan en bins: rangos equidistantes de agrupaciones de valores numéricos.



Son usualmente el primer gráfico a realizar de una columna numérica.







El eje Y siempre debe empezar en cero.








No tiene sentido hacer un histograma de una variable categórica. Esto sería un diagrama de barras.


## Objetivo:
Analizar la distribución de edades de los pasajeros del Titanic.

```python
import pandas as pd
```

```python
df = pd.read_csv('./titanic.csv')
df = df.dropna(subset = ['Age'], axis=0)
df
```

```python
df.hist(
    'Age',
    bins = 5 #Muy pocos
)
```

```python
df.hist(
    'Age',
    bins = 200 #Excesivo, dificil de entender
    
)
```

```python
df.hist(
    'Age',
    by = 'Sex',
    bins = 15
)
```

# KDE

Kernel density estimator

Es un suavizado del histograma que tiene la forma de la distribución de los valores numéricos. Formalmente es una función de probabilidad.









```python
import seaborn as sns
```

```python
sns.histplot(
    df,
    x='Age',
    kde = True,
    bins = 20,
    stat = 'probability',
#     hue = 'Sex' #integrar variable categorica al histograma
)
```

```python
sns.histplot(
    df,
    x='Age',
    cumulative = True,
    bins = 20,
    fill = False,
    element = 'step'
    
    #Escalon mas largo donde hay mas datos, histograma acumulativo
)
```
