```python
import pandas as pd
```

```python
df = pd.read_csv('./titanic.csv')
df
```

### ¬øCu√°ntos pasajeros por sexo tuvo el Titanic?
#### Objetivo:
Visualizar cuantos pasajeros hombres y mujeres habian en el Titanic

```python
pasajeros = df.groupby('Sex').agg({
    'PassengerId': 'count'
})
pasajeros
```

```python
#pandas usa matplotlib para graficacion

pasajeros.plot.bar(
    y = 'PassengerId',
    width = 0.3,
    title = 'Numero de pasajeros por sexo'
)
```

```python
import seaborn as sns
```

```python
pasajeros2 = pasajeros.reset_index()
pasajeros2
```

```python
sns.barplot(
    data = pasajeros2,
    x='Sex',
    y='PassengerId'
)
```
