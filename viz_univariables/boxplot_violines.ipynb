# Grficos de cajas

Muestra estadsticos de una variable de uno o ms grupos de mis datos mediante cajas y lneas









## Cundo utilizar?

Quiero comparar diversos grupos a travs de sus estadsticos (media, mediana, cuartiles).

Quiero detectar datos aberrantes de mi variable o dentro de alguna categora


## Cundo no utilizar?

Si me interesa observar la forma exacta de mi distribucin.

Si quiero analizar los valores individuales de mi variables. En su forma ms simple, el diagrama de cajas no me ayuda para esto.

## Consejos

Es indiferente hacer las cajas verticales u horizontales



```python
import pandas as pd
```

```python
df = pd.read_csv('./titanic.csv')
df
```

```python
df = df.dropna(subset=['Age'], axis='index') #elimina la fila
```

```python
df.boxplot(
    column = 'Age',
    by = 'Sex'
)
```

```python
import seaborn as sns
```

```python
sns.boxplot(
    data = df,
    x='Sex', #Categorica
    y='Age' #Numerica
)
```

# Grfico de violines

Manteniendo todas las ventajas de un diagrama de cajas, me muestra la forma de distribucin de mi variable.









## Cundo utilizar?

Quiero ver o comparar la forma de mi distribucin entre diferentes grupos y al mismo tiempo comparar sus estadsticos.

## Cundo no utilizar?

Si no me interesa la forma de mi distribucin, puedo conformarme con un simple diagrama de cajas.



```python
sns.violinplot(
    data = df,
    x='Sex',
    y='Age',
    inner = 'quartiles' #stick-barra por cada dato, point
)
```
