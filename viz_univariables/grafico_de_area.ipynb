# Gráfico de área o area chart






El eje Y sí empieza en cero para poder diferenciar magnitudes.









```python
import pandas as pd
```

```python
df = pd.read_csv('./tienda_ventas.csv')
df = df.set_index('id')
df
```

```python
df['mes'] = df['date'].apply(lambda x: x[:7])

df_por_mes = df.groupby('mes').agg({
    'sales':'sum'
    })
df_por_mes
```

```python
df_por_mes.plot.area(
    y='sales'
)
```
