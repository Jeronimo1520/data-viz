### Filtrado por condiciones

```python
df[df['full_text'].str.contains('Programming')]
```

```python
def calcularGanancias(retweets):
    ganancias = retweets * random.randint(3, 5)
    return ganancias

df_filtrado3['ganancias'] = df_filtrado3['retweets'].apply(calcularGanancias)
df_filtrado3.head(10)
```

```python
def popularidad(fila):
    resultado = fila['followees'] / fila['followers']
    return resultado

df['popularidad'] = df.apply(popularidad, axis=1)
#Axis=1 especifica a .apply() que la funcion debe ser aplicada por cada fila del df
df.head()
```

### Filtrado por agrupacion

```python
df[ 
    ['country', 'favorites', 'retweets', 'mentions', 'followers', 'followees', 'ganancias', 'popularidad']
].groupby("country").mean()
```

```python
grouped = df[ 
    ['country', 'favorites', 'retweets', 'mentions', 'followers', 'followees', 'ganancias', 'popularidad']
].groupby("country").agg({
    'followers':'sum',
    'mentions':'mean',
    'retweets':'max'
})

grouped[grouped['followers'] > 5000]
```

### Guardado de datos 

```python
grouped.to_csv("salida.csv")
```
