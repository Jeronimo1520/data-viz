# Gráfico de Dispersión (Scatter Plot)

Gráfico que muestra los valores de dos variables de un mismo registro en nuestro dataset al mismo tiempo. En cada Eje del gráfico se coloca una variable. Se dibuja en el canvas un punto, en la coordenada correspondiente al valor de las variables por cada observación de nuestro dataset. Su objetivo es detectar si existe o no alguna correlación entre las variables en los ejes. Estos ejes usualmente son variables numéricas pero también se pueden utilizar variables categóricas ordinales.

**Eje X**: Variable numérica o categórica,
**Eje Y**: Variable numérica o categórica,
**Otros nombres**: Gráfico de puntos



**Ejemplos de Uso**:

- Como un doctor, me gustaría saber si el nivel de colesterol de una persona se correlaciona con su peso.
