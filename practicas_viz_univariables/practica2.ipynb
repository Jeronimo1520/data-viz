```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./datasets_top_4.csv')
df
```

```python
fig, axes = plt.subplots(1, 2)
```

```python
fig, (ax1, ax2) = plt.subplots(
    1, # Numero de filas
    2, # Numero de Columnas
    constrained_layout=True # Permite que los subgraficos dejen espacio para los titulos y los ejes
)
fig.set_size_inches(10, 5)
df1 = df[0:13]
#df1 = df.iloc[0:13, :] #Equivalente, se cojen 13 filas y todas las columnas
df2 = df[13:]
#df2 = df.iloc[13:, :] #Equivalente, se cojen las filas restantes despues de 13 y todas las columnas
fig.suptitle('CÃ³mo se distribuyen las categorÃ­as dentro de los datasets?', size = 15)
color = []
df1.plot.barh(stacked = True, ax = ax1, width = 0.2,color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)

ax1.set_xlabel('Porcentaje')
ax1.set_ylabel('Dataset')
ax1.set_yticklabels(
    df1['dataset'], 
)

fig.text(
    0.4, -0.15,
    "Algunos datasets han sido ignorados del analisis",
    fontsize=6,
    bbox={
        "facecolor": "orange",
        "alpha": 0.5
    }
)
ax1.set_xticks(range(0, 101, 25))

df2.plot.barh(stacked = True, ax = ax2, width = 0.2,color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)

ax2.set_xlabel('Porcentaje')
ax2.set_yticklabels(
    df2['dataset'], 
)
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
ax2.tick_params(axis='y', labelright=True)

ax2.set_xticks(range(0, 101, 25))

ax1.legend().set_visible(False)
ax2.legend().set_visible(False)

labels = ['Categoria 1', 'Categoria 2', 'Categoria 3', 'Otros']

fig.legend(labels,
           loc='lower center',
           ncol=len(labels),
           bbox_to_anchor=(0.5, -0.1),
          )
```
