```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./viajes_surfistas.csv')
df
```

```python
paises_definidos = ['United States of America', 'Australia', 'Japan', 'Spain']
df['homecountryparagraficar'] = np.where(df['homecountry'].isin(paises_definidos), df['homecountry'], 'Otros')
df_pie = df.groupby('homecountryparagraficar').count()
df_pie = df_pie.sort_values(by='homecountry', ascending=False)
df_pie
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
colors = plt.cm.Pastel1(range(len(df_pie))) 

df_pie.plot.pie(
    y = 'homecountry',
     autopct='%1.1f%%',
    startangle = 290,
    colors = colors,
    ax = ax
    
)

ax.set_ylabel('')  # Eliminar el tÃ­tulo del eje Y
ax.set_title('Porcentaje de Surfistas por PaÃ­s de Procedencia', fontsize=16)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), title='PaÃ­s de surfistas')
plt.show()
```

```python
agrupado = df.groupby('homecountry').agg({
    'homename': 'count' 

})
```

```python
#Ordenar el df
agrupado_ordenado = agrupado.sort_values('homename', ascending = False)
```

```python
top_4 = agrupado_ordenado.head(4)
top_4
```

```python
df_otros = agrupado_ordenado.drop(index=top_4.index.values) #Eliminar top 4 paises
df_otros
```

```python
df_otros['pais'] = 'Otros'
df_otros
```

```python
df_otros_agrupado= df_otros.groupby('pais').agg({
    'homename' : 'sum'
})
df_otros_agrupado #Obtener suma de los otros paises
```

```python
df_otros_agrupado = df_otros_agrupado.reset_index()
df_otros_agrupado #Volver columnas los indices predeterminados y volver idx numerico
```

```python
top_4 = top_4.reset_index()
top_4
```

```python
#Unir ambos df para graficar
top_4.columns = ['pais','homename'] #Las columnas deben ser las mismas en ambos df
top_4
```

```python
df_final = pd.concat([top_4,df_otros_agrupado ]) #Concatenar dfs
df_final
```

```python
#Para graficar setear el indice del df para que sea el pais
df_final = df_final.set_index('pais')
df_final
```

```python
ax = df_final.plot.pie(
    y = 'homename',
    startangle = 90,
    autopct = '%1.1f%%',
    colormap = 'Pastel1'
)

ax.set_ylabel('')
ax.legend(
    title = "PaÃ­s de surfistas",
    fontsize = 8,
    bbox_to_anchor = (1,1)
)

ax.set_title("Porcentaje de Surfistas por PaÃ­s de Procedencia")
```
