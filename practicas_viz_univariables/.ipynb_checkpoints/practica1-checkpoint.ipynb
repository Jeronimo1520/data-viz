```python
import pandas as pd
```

```python
df = pd.read_csv('./comparasion_algoritmos.csv')
df
```

```python
ax = df.plot.bar(
    y = ['algoritmo1','algoritmo2'],
    width = 0.5,
    title = 'Evaluación de Algoritmo1 vs Algoritmo2',
    figsize = (12, 5) #Ancho y alto de la figura en pulgadas, sobreescribe el global
    
)

ax.set_xticklabels(
    df['dataset']
)

ax.legend(
    labels = ['Algoritmo 1', 'Algoritmo 2']
)

ax.set_xlabel(
    'Dataset',
    loc = 'center'
)

ax.set_ylabel(
    'Rendimiento',
    loc = 'center'
)

ax.grid(True)
ax.set_axisbelow(True) #Poner lineas del grid atras de las barras
```
